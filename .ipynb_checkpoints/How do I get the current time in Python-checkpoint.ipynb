```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python"
response = requests.get(url)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
question_title = soup.find('a', class_='question-hyperlink').get_text()
question_body = soup.find('div', class_='s-prose js-post-body').get_text()
```

```python
answers = soup.find_all('div', class_='answer')
answer_texts = [answer.find('div', class_='s-prose js-post-body').get_text() for answer in answers]
```

```python
print("Question Title:", question_title)
print("\nQuestion Body:\n", question_body)
print("\nAnswers:\n")
for idx, answer in enumerate(answer_texts):
    print(f"Answer {idx + 1}:\n{answer}\n")
```
